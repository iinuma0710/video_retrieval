```python
import os
import cv2
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_dir = "/net/per610a/export/das18a/satoh-lab/share/datasets/eastenders/trecvid_2019/topics/tv19.person.examples.images/"
mask_images = glob.glob(os.path.join(image_dir, "*.mask.png"))
```

```python
mask_img_path = mask_images[5]
src_img_path = mask_img_path.replace("mask", "src")
```

```python
mask_im = cv2.imread(mask_img_path, cv2.IMREAD_GRAYSCALE)
indexes = np.where(mask_im == 255)
top = np.min(indexes[0])
bottom = np.max(indexes[0])
left = np.min(indexes[1])
right = np.max(indexes[1])
```

```python
top, bottom, left, right
```

```python
src_im = cv2.imread(src_img_path)
trimmed_img = src_im[top:bottom+1, left:right+1]
```

```python
plt.imshow(cv2.cvtColor(trimmed_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
for mask_img_path in mask_images:
    src_img_path = mask_img_path.replace("mask", "src")
    trimmed_img_path = mask_img_path.replace("mask", "trimmed")
    mask_im = cv2.imread(mask_img_path, cv2.IMREAD_GRAYSCALE)
    indexes = np.where(mask_im == 255)
    top = np.min(indexes[0])
    bottom = np.max(indexes[0])
    left = np.min(indexes[1])
    right = np.max(indexes[1])
    src_im = cv2.imread(src_img_path)
    trimmed_img = src_im[top:bottom+1, left:right+1]
    cv2.imwrite(trimmed_img_path, trimmed_img)
```
