```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
video_path = "/net/per610a/export/das18a/satoh-lab/share/datasets/eastenders/video_detected/shot95_1455_id_1050566.mp4"
```

```python
cap = cv2.VideoCapture(video_path)

frames_0 = []
while True:
    ret, frame = cap.read()
    if ret:
        frames_0.append(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    else:
        break
```

```python
video_path = "/net/per610a/export/das18a/satoh-lab/share/datasets/eastenders/video_detected/shot35_1514_id_383003.mp4"
```

```python
cap = cv2.VideoCapture(video_path)

frames_3 = []
while True:
    ret, frame = cap.read()
    if ret:
        frames_3.append(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    else:
        break
```

```python
frames = []
for fs in [frames_0, frames_1, frames_2, frames_3]:
    interval = int((len(fs) - 1) / 4)
    frames += [fs[i] for i in range(0, len(fs), interval)]
```

```python
row = 4
col = 5
plt.figure(figsize=(16,12))

num = 0

while num < row * col:
    num += 1
    plt.subplot(row, col, num)
    plt.imshow(frames[num - 1])
    if num % 5 == 1:
        label = "({})".format(int((num - 1) / 5 + 1))
        plt.ylabel(label, rotation=0, fontsize=20, labelpad=50)
    ax = plt.gca()
    ax.axes.xaxis.set_ticks([])
    ax.axes.yaxis.set_ticks([])
#     plt.axis('off')
```
