```python
with open("../data/eastenders/grand_truth.csv", "r") as fp:
    reader = csv.reader(fp, delimiter=' ')
    action_label_dict = {r[0]: int(r[1]) for r in reader}
    
with open("../data/eastenders/original_videos.csv", "r") as fp:
    reader = csv.reader(fp, delimiter=' ')
    original_video_dict = {os.path.basename(v[1])[:-4]: [v[0], v[2]] for v in reader}
    
with open("../data/eastenders/person_videos.csv", "r") as fp:
    reader = csv.reader(fp, delimiter=' ')
    person_video_dict = {os.path.basename(v[2])[:-4]: [v[0], v[3]] for v in reader}
```

```python
feature_id = 1
features_csv = []
action_features = []
for idx in np.unique(label[:, 1]):
    # 
    fv = np.mean(feature[np.where(label[:, 1] == idx)], axis=0)
    # 
    video_path = video_index_dict[idx]
    video_file = re.split("[_.]", os.path.basename(video_path))
    video_shot = video_file[0] + "_" + video_file[1]
    person_shot = video_file[0] + "_" + video_file[1] + "_id_" + video_file[3]
    # original_video_id, action_gt_label
    original_video_id, action_gt_label = original_video_dict[video_shot]
    # person_video_id, person_gt_label
    person_video_id, person_gt_label = person_video_dict[person_shot]
    # 
    action_features.append(fv)
    features_csv.append([feature_id, original_video_id, person_video_id, feature_id-1, None, video_path, action_gt_label, person_gt_label])
    feature_id += 1
```

```python
if os.path.exists("../data/eastenders/features.csv"):
    with open("../data/eastenders/features.csv", "a") as fp:
        writer = csv.writer(fp, delimiter=' ')
        writer.writerows(features_csv)
else:
    with open("../data/eastenders/features.csv", "w") as fp:
        writer = csv.writer(fp, delimiter=' ')
        writer.writerows(features_csv)
        
action_features = np.array(action_features)
if os.path.exists("../data/eastenders/action_features.npy"):
    fvs = np.load("../data/eastenders/action_features.npy")
    fvs = np.concatenate([fvs, action_features])
    np.save("../data/eastenders/action_features.npy", fvs)
else:
    np.save("../data/eastenders/action_features.npy", action_features)
```

```python
feature_id = 1

for i in range(12):
    feature = np.load("../data/eastenders/feature_label_32/feature_{}.npy".format(i))
    label = np.load("../data/eastenders/feature_label_32/label_{}.npy".format(i))
    
    features_csv = []
    action_features = []
    for idx in np.unique(label[:, 1]):
        # 
        fv = np.mean(feature[np.where(label[:, 1] == idx)], axis=0)
        # 
        video_path = video_index_dict[idx]
        video_file = re.split("[_.]", os.path.basename(video_path))
        video_shot = video_file[0] + "_" + video_file[1]
        person_shot = video_file[0] + "_" + video_file[1] + "_id_" + video_file[3]
        # original_video_id, action_gt_label
        original_video_id, action_gt_label = original_video_dict[video_shot]
        # person_video_id, person_gt_label
        person_video_id, person_gt_label = person_video_dict[person_shot]
        # 
        action_features.append(fv)
        features_csv.append([feature_id, original_video_id, person_video_id, feature_id-1, None, video_path, action_gt_label, person_gt_label])
        feature_id += 1
        
    if os.path.exists("../data/eastenders/features.csv"):
        with open("../data/eastenders/features.csv", "a") as fp:
            writer = csv.writer(fp, delimiter=' ')
            writer.writerows(features_csv)
            print("Add features data.", i)
    else:
        with open("../data/eastenders/features.csv", "w") as fp:
            writer = csv.writer(fp, delimiter=' ')
            writer.writerows(features_csv)
            print("Create features data.", i)

    action_features = np.array(action_features)
    if os.path.exists("../data/eastenders/action_features.npy"):
        fvs = np.load("../data/eastenders/action_features.npy")
        fvs = np.concatenate([fvs, action_features])
        np.save("../data/eastenders/action_features.npy", fvs)
    else:
        np.save("../data/eastenders/action_features.npy", action_features)
```
